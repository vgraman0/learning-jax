```python
import numpy as np
from jax import jit
from jax import lax
from jax import random
import jax
import jax.numpy as jnp
```

```python
def impure_function(x):
    print("Executing function")
    return x

print("First call: ", jit(impure_function)(4.))
print("Second call: ", jit(impure_function)(5.))
print("Third call: ", jit(impure_function)(jnp.array([5.])))
```

```python
g = 0.
def impure_globals(x):
    return x + g

print("First call: ", jit(impure_globals)(4.))
g = 10.

print("Second call: ", jit(impure_globals)(5.))

print("Third call: ", jit(impure_globals)(jnp.array([5.])))
```

Impure functions won't behave as expected do to caching. 

```python
from jax import make_jaxpr

def permissive_sum(x):
    return jnp.sum(jnp.array(x))

x = list(range(10))
make_jaxpr(permissive_sum)(x)
```
