# What is stored in memory when you write x = 5?

A **name** (x) is bound to an object (the integer 5) that lives on the heap. Each code cell below probes one part of that model.

```python
x = 5
type(x)
```

```python
hex(id(x))
```

First, note that x is an `int` with a hexadecimal address. This is the location of the integer 5 in memory.

```python
y=4
y = y + 1
x is y
```

```python
c = 257
d = 257
c is d
```

CPython has a small-int cache for the values -5 to 256. These are simply the integers that show up the most often.

Small-ints share the cached object.

```python
import sys

nums = [-7, 4, 0, 5, 2**30, 2**60]
for num in nums:
    print(num, sys.getsizeof(num))
```

```python
print(int.__bases__)        # (<class 'object'>,)
print(isinstance(5, object))  # True — everything subclasses object
print(type(5))              # <class 'int'>
print(type(int))            # <class 'type'>
```
